```python
import numpy as np
```

```python
# Basic attributes of a numpy array.
```

```python
#Create a numpy array from a list
l1 = [1, 2, 3, 4, 5]
arr = np.array(l1)
print(type(arr))    # To print the object type
print(arr.ndim)     # Prints the dimension the the ndarray
print(arr.dtype)          # Prints the datatype of the value sstored in the array
print(arr.size)            # Prints the number of elements in the array
print(arr.nbytes)          # Prints the total size of the array in bytes
print(arr.itemsize)        # Prints the size of an individual element in the array
```

```python
# Indexing an element in a single dimensional array.
# Syntax: array[index_value]
```

```python
print(arr[0]) # Prints the value at the index 0
print(arr[3]) # prints the value at the index 3
```

```python
# Slicing a single dimensional array.
# Syntax: array[start:stop:step]
```

```python
print(arr[1:3])     # Prints the values from index 1 inclusive to 3 exclusive.
print(arr[::])      # Prints all the values in the array.
print(arr[:4])      # Prints all the values from index 0 to 4 exclusive.
print(arr[::2])     # Prints all the even indexed elements.
print(arr[5:2:-1])  # Prints the values from 5 inc to 2 exclusive using negative indexing.
print(arr[::-1])    # Prints the values in the array in the reverse order.
```

```python
# Multidimensional arrays.
```

```python
arr2 = np.array([[1, 2, 3, 4], [5, 6, 7, 8]])
print(arr2.ndim)
print(arr2.shape)  # Returns the shape in the form of (rows, columns)
arr2.reshape(4, 2) # Reshape the array into a 4x2 array.(No copy is created).

# Indexing
print(arr2[0, 2])
print(arr2[0, 0])
print(arr2[1][3])

# craeting copies of an array.
arr3 = arr2.copy()
arr3              # It is a copy of the array arr2
arr4 = arr.copy()
arr4              # It is the copy of the array arr
```

```python
# Concatenating 2 or more arrays.
# Syntax: np.concatenate([array1, array2, array3....], axis=)
```

```python
a1 = np.array([1, 2, 3, 4])
a2 = np.array([5, 6, 7, 8])
np.concatenate([a1, a2], axis=0)

# hstack or horizontal stack
np.hstack([a1, a2])

# Vertical Stack
b = np.vstack([a1, a2])
```

```python
# Splitting arrays
#Syntax: np.split(array_name, [split_points], axis=)
```

```python
a1 = np.array([1, 2, 3, 4])
a2 = np.array([5, 6, 7, 8])
b = np.vstack([a1, a2])
np.split(b, [1, 2], axis=1)

#hsplit
np.hsplit(b, 2)

#vsplit
np.vsplit(b, 2)
```

```python
# Some important numpy functions
```

```python
arr = np.array([1, 2, 3, 5, 4])
arr2 = np.array([0.122, 0.143, 0.655, 0.878, 0.644, 0])
print(np.min(arr))
print(np.max(arr))
print(np.linspace(0, 10, 11)) # Prints 11 values b/w 0 and 10 with equal intervals.
print(np.logspace(1, 5, 10))  # The logspace() function return numbers spaced evenly on a log scale.
print(np.zeros(9).reshape(3, 3))  # Prints an array with all zeros
print(np.ones(9).reshape(3, 3))   # Prints an array with all ones
print(np.sin(arr))                # Prints the sin of all the array values

# Mathematical functions.
print(np.cos(arr))
print(np.arcsin(arr2))            # Answer is in terms of radians
print(np.sum(arr2))
print(np.prod(arr))
```
