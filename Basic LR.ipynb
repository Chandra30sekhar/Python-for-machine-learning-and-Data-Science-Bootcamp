```python
import numpy as  np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('./basic LR/test.csv')
```

```python
df
```

```python
x = df['x']
y = df['y']
```

```python
plot = plt.scatter(x, y)
```

```python
m, c = np.polyfit(x, y, deg=1)
```

```python
test_data = pd.read_csv('./basic LR/test.csv')
```

```python
test_data_x = test_data['x']
test_data_y = test_data['y']
```

```python
pred_values = []
for val in test_data_x:
    pred_values.append((m*val)+c)
```

```python
for i in range(len(test_data_y)):
    print(f"actual: {test_data_y[i]}, obtained: {pred_values[i]}")
```

```python
# Reduce the error using rms
```

```python
actual_count = len(test_data_y)
count = 0
for i in range(actual_count):
    if test_data_y[i] == pred_values[i]:
        count += 1
```
