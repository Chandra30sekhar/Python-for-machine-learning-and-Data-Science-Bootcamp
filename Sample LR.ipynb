# Import Required Packages

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
x = np.linspace(1, 100, num=100)
y = 12 * x + 19
```

# Split training and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y)
```

# Plot basic graphs

```python
plt.scatter(X_train, Y_train) #using matplotlib
```

```python
sns.distplot(X_train, kde=False)
```

```python
sns.jointplot(x=X_train, y=Y_train) #using seaborn
```

# Create and train LR model

```python
m, c = np.polyfit(X_train, Y_train, deg=1)
m, c # 18.99 due to Gaussian Error
```

```python
plt.plot(m*X_test+c, Y_test)
```

# Model Evaluation

```python
from sklearn import metrics
```

```python
y_pred = m*X_test + c        
```

```python
print("MSE: ", int(metrics.mean_squared_error(Y_test, y_pred)))
```
