```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
#note: use plots based on audience.
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

# Distribution Plots

```python
sns.distplot(tips['total_bill'], kde=True) 
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
 sns.pairplot(tips, hue='sex', palette='coolwarm')
```

```python
sns.rugplot(tips['total_bill'])
```

# Categorical Plots
used for categorical data

```python
sns.barplot(x='sex', y='total_bill', data=tips)
```

```python
sns.countplot(x='sex', data=tips) #estimator = count of each object
```

```python
sns.boxplot(x='day', y='total_bill', data=tips) #used to show distribution of categorical data
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True) #shows more information than box plot
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex', split=True)
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips, hue='sex') #scatter and strip plot
#drawback: not scalable, more computation time required.
```

```python
sns.factorplot(x='day', y='total_bill',data=tips) # more genral form
```

# Matrix Plots

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
tc = tips.corr()
```

```python
sns.heatmap(tc, annot=True, cmap='coolwarm')
```

```python
fp = flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
sns.heatmap(fp, cmap='magma', linecolor='white', linewidth=1)
```

```python
# clustermap
# uses hierarchical clustering
# clusters data such that similar groups are closer to each other.
sns.clustermap(fp, cmap='magma', standard_scale=1)
```

```python
#grids
#automate subplots based on features
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(iris)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
g = sns.FacetGrid(data=tips, col='time',row='smoker')
g.map(sns.distplot, 'total_bill')
```

```python
#regression plots
#lm plot is used to plot linear functions
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o', 'x'],
           scatter_kws={'s':100})
#NOTE: calls matplotlib
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='smoker', row='time', hue='sex',
          aspect=0.6, height=8)
```
